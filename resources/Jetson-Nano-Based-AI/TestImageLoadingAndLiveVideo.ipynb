```python
import cv2
from IPython.display import display, Image

var = 1 # 0 = camera foto, 1 = geuploade foto, 2 = live camera feed

if var == 0:
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cap.release()

    if ret:
        print("Camera werkt!")
        _, jpeg = cv2.imencode(".jpg", frame)
        display(Image(data=jpeg))
    else:
        print("Kan geen frame lezen...")
    
elif var == 1:
    image_path = "./Verkeersbord.jpg"  # bestand in Jupyter
    img = cv2.imread(image_path)
    _, jpeg = cv2.imencode(".jpg", img)
    display(Image(data=jpeg))

elif var == 2:
    import time
    from IPython.display import clear_output

    cap = cv2.VideoCapture(0)

    start_time = time.time()
    duration = 10  # seconds

    if not cap.isOpened():
        print(" Cannot open camera")
    else:
        print(" Live camera feed (10 seconds)")

        while time.time() - start_time < duration:
            ret, frame = cap.read()
            if not ret:
                print(" Failed to grab frame")
                break

            _, jpeg = cv2.imencode(".jpg", frame)

            clear_output(wait=True)
            display(Image(data=jpeg))

            time.sleep(0.05)  # ~20 FPS, prevents CPU overload

    cap.release()
    print(" Camera released")

else:
    print("Geen methode geselecteerd")
```

```python
import cv2
from IPython.display import display, Image

#benodigde variabelen:
image_path = "./Verkeersbord.jpg"  # bestand in Jupyter
fps = 50 # aantal verwachte frames per second van de camera
var = 2 # 0 = camera foto, 1 = geuploade foto, 2 = live camera feed
duration = 10  # seconden dat de camera aanstaat

if var == 0:
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cap.release()

    if ret:
        print("Camera werkt!")
        _, jpeg = cv2.imencode(".jpg", frame)
        display(Image(data=jpeg))
    else:
        print("Kan geen frame lezen...")
    
elif var == 1:
    img = cv2.imread(image_path)
    _, jpeg = cv2.imencode(".jpg", img)
    display(Image(data=jpeg))

elif var == 2:
    import time
    from IPython.display import clear_output

    cap = cv2.VideoCapture(0)

    start_time = time.time()

    if not cap.isOpened():
        print(" Cannot open camera")
    else:
        print(" Live camera feed (10 seconds)")

        while time.time() - start_time < duration:
            ret, frame = cap.read()
            if not ret:
                print(" Failed to grab frame")
                break

            _, jpeg = cv2.imencode(".jpg", frame)

            clear_output(wait=True)
            display(Image(data=jpeg))

            time.sleep(0.01)  # ~20 FPS, prevents CPU overload

    cap.release()
    print(" Camera released")

else:
    print("Geen methode geselecteerd")
```
