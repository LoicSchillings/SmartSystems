```python
import cv2
from IPython.display import display, Image

cap = cv2.VideoCapture(0)
ret, frame = cap.read()
cap.release()

if ret:
    print("Camera werkt!")
    _, jpeg = cv2.imencode(".jpg", frame)
    display(Image(data=jpeg))
else:
    print("Kan geen frame lezen...")
```
